```python
# imports and housekeeping
import numpy as np
import pandas as pd
import csv
import os
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from itertools import imap, chain
from operator import sub
from scipy.stats.stats import pearsonr
from os import listdir
from os.path import isfile, join
import pylab

%matplotlib inline
```

```python
# Plot all Archetype usage between 8/7/16 and 6/11/17

HS_archs = os.listdir('Z:/ELSA/Data/Archetype by Class')

for files in HS_archs:
    tmp = pd.read_csv('Z:/ELSA/Data/Archetype by Class/' + files, sep = ',', header = 0);
    tmpfig = tmp.plot(figsize = (9,6), xticks = range(43));
    tmpfig.set_xticks(range(43));
    tmpfig.set_xticklabels(tmp['x']);
    tmpfig.set_xticklabels(tmpfig.xaxis.get_majorticklabels(), rotation=70);
    tmpfig.set_title(files);
```
